```python
import joblib
import pandas
import gensim
import numpy as np
import pandas as pd
import re
import gensim.corpora as corpora
from gensim.test.utils import datapath
from gensim.models import LdaModel
from pprint import pprint
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
path = datapath("/Users/wvdvijve/Documents/GitHub/becode-projects/KPMG_NLP_project/models/mallet")

model = LdaModel.load(path)

pattern = r'\"([\w]*)\"'
pat = re.compile(pattern)
topics = {}
for topic in model.print_topics():
    new_text = pat.findall(topic[1])
    topics[topic[0]] = new_text
```

# The model

```python
topics_df = pd.Series(topics, name='keywords')
topics_df
```
